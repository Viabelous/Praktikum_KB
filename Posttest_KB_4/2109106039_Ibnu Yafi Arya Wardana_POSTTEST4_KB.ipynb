## Data Cleaning (NULL Remover)

Lakukan Data cleaning pada data dengan nilai null (jika tidak ada nilai null pada dataset, maka buatlah menjadi ada). Ganti nilai null sesuai ketentuan. (bilangan bulat : median/modus, bilangan desimal : mean, tulisan : modus). Nilai plus jika strategy yang digunakan menggunakan 3 nilai

```python
df = pd.read_csv("phones_data.csv")
df.isna().sum()
```

```python
from sklearn.impute import SimpleImputer
```

##### Modus:

```python
imputer = SimpleImputer(strategy='most_frequent')
df['os'] = imputer.fit_transform(df[['os']])
```

#### Median:

```python
imputer = SimpleImputer(strategy='median')
df['sellers_amount'] = imputer.fit_transform(df[['sellers_amount']])
```

#### Mean:

```python
imputer = SimpleImputer(strategy='mean')
df[['lowest_price',
    'highest_price',
    'memory_size',
    'screen_size',
    'battery_size']] = imputer.fit_transform(df[['lowest_price',
                                                 'highest_price',
                                                 'memory_size',
                                                 'screen_size',
                                                 'battery_size']])
```
