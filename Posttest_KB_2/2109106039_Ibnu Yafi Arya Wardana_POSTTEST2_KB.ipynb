#### Ketentuan data analysis 2
Masukkan informasi di atas dalam **dictionary** dengan **key sesuai nama informasinya**. Tampilkan isi dari dictionary tersebut sebagus mungkin.

**Jawab:**  
1. Masukkan informasi ke dalam dictionary
2. Tampilkan key dan value satu-persatu menggunakan for

```python
dictionaries = {
    'Jumlah Record'  : len(df.values),
    'Nama Atribute'  : [i for i in df.columns],
    'Atribut Object' : len(df.select_dtypes(include = ['object']).columns),
    'Atribut Angka'  : len(df.select_dtypes(include = ['float64', 'int64']).columns),
    'Atribut Null'   : len(df.columns[df.isnull().any()])
}

print("   Informasi\t|             Jumlah")
print("*"*50)

for key,value in dictionaries.items():
    if key != 'Nama Atribute':
        print(f"{key} \t|  {value}")
        
    else:
        print(f"{key} ",end="\t| ")
        j = 0

        for i in value:
            j += 1
            print(f" {j}. {i}")
            
            if j != len(value):
                print("\t\t| ",end="")
                
    print("-"*50)
```

#### Ketentuan data analysis 5
**Rubah nilai** salah satu attribute bertipe angka menjadi **desimal 3 angka dibelakang koma** jika tidak ada, lakukan sebaliknya (dari desimal menjadi bilangan bulat).

**Jawab:**  
Untuk mengubah nilai sementara bisa dengan menggunakan .astype(tipe data)  
**akan tetapi** untuk mengganti secara "selamanya" harus dengan menindih  
data (  df = df.astype()   )

```python
list=[]
for col in df.columns:
    if df[col].dtypes == 'int64':
        list.append(col)
    

df[list] = df[list].astype('float64').applymap("{:.3f}".format, "ignore")

display(df)
```
