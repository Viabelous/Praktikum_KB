## Import Library

```python
import pandas as pd
from matplotlib import pyplot as plt
# import matplotlib as plt
import seaborn as sns
```

## Read Data

```python
df = pd.read_csv('marketplace.csv')
```

# Ketentuan 1
Gunakan minimal 3 tipe visualiasi data  
*nb: wajib menggunakan pie chart  
*jelasin yang terlihat pakai markdown

**----------Pie Chart--------**

```python
data_pie = df['merchant_badge'].value_counts().rename_axis("Badge Penjual").reset_index(name="Banyak")
# desc:
# value_counts = hitung jumlah
# rename_axis  = kasih nama kategori
# reset_index  = kasih nama jumlahnya

list = [i for i in df['merchant_badge'].value_counts()]

plt.pie(
    data_pie["Banyak"],
    labels     = data_pie["Badge Penjual"],
    startangle = 90,                       #<-- perotasi
    autopct    = '%.2f%%',                 #<-- penampil angka
    explode    = (0.1,0.1,0),              #<-- pemisah antar kategori
    shadow     = True,                     #<-- efek 3D aka. bayangan
    colors     = ['#66b3ff','#ffcc99','#99ff99'] #<-- warna
)

plt.title("Pendataan Berdasarkan Badge Penjual ",
          bbox={'facecolor':'0.9', 'pad':5})

plt.text(1.6,-0.3, f" Banyak data PM: {list[0]}\n\
                 \n Banyak data OS: {list[1]}\n\
                 \n Banyak data RM: {list[2]}"
         , bbox={'facecolor':'0.9', 'pad':5})

plt.show()
```

Pada data di atas dapat dilihat:  
Penjual dengan Badge PM adalah yang paling banyak terdata dengan besar bagian: 78.24% dengan 20353 data
Penjual dengan Badge RM terdata dengan besar bagian: 12.62% dengan 3284
Penjual dengan Badge OS adalah yang paling sedikit terdata dengan besar bagian: 9.14% dengan 2377

**----------Pie Chart--------**

```python
df['date'] = pd.to_datetime(df["date"])

#Resample jadi 'M' aka. Month
rate_monthly = df.resample('m', on='date').mean()

#Ngebesarin Ukuran Garis (Sunnah)
plt.figure(figsize=(10,10))

#Tampilkan
sns.lineplot(
    data = rate_monthly,
    x    = rate_monthly.index,
    y    = 'rating'
)

plt.title ("Kenaikan Rating Berdasarkan Waktu")
plt.show()
```

Pada visualisasi di atas dapat terlihat, seiring berjalannya waktu, rating yang didapat juga terus meningkat
